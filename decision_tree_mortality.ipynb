```python
import pandas as pd

df = pd.read_csv('IHME_GBD_2010_MORTALITY_AGE_SPECIFIC_BY_COUNTRY_1970_2010.csv')
df['Death Rate Per 100,000'] = df['Death Rate Per 100,000'].replace(',', '', regex=True).astype(float)
df = df.dropna(axis=0)
```

```python
df['Sex_n'] = df['Sex'].astype('category').cat.codes
df['Age_n'] = df['Age Group'].astype('category').cat.codes
df['Country_n'] = df['Country Name'].astype('category').cat.codes
```

```python
y = df['Death Rate Per 100,000']
features = ['Year', 'Sex_n', 'Age_n', 'Country_n']
X = df[features]

from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

tree_model = DecisionTreeRegressor(random_state=1)
tree_model.fit(train_X, train_y)
tree_mae = mean_absolute_error(val_y, tree_model.predict(val_X))

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
forest_mae = mean_absolute_error(val_y, forest_model.predict(val_X))

print(f"Ошибка дерева: {tree_mae}")
print(f"Ошибка леса: {forest_mae}")
```
